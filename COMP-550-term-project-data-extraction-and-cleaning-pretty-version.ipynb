### 20 Newsgroups Data Extraction
#### To exclude the information in the files' headers, we verified that every file contained a line starting with "Subject:" near the end of their header:

```python
root_dir = '\\\\?\\' + os.path.abspath(os.path.join(os.getcwd(), '20_newsgroups'))
file_count = 0
missing_subject_line_count = 0
total_files_count = 0

subdirectory_list = ['alt.atheism', 'comp.graphics', 'comp.os.ms-windows.misc', 'comp.sys.ibm.pc.hardware', 'comp.sys.mac.hardware', 'comp.windows.x', 'misc.forsale', 'rec.autos', 'rec.motorcycles', 'rec.sport.baseball', 'rec.sport.hockey', 'sci.crypt', 'sci.electronics', 'sci.med', 'sci.space', 'soc.religion.christian', 'talk.politics.guns', 'talk.politics.mideast', 'talk.politics.misc', 'talk.religion.misc']

for subdir in subdirectory_list:
    
    # making a list of relevant files
    files = [ os.path.join(root_dir, subdir, f) for f in os.listdir(os.path.join(root_dir, subdir)) if '.onetoc2' not in f ]
    
    total_files_count += len(files)
    
    # searching for "landmark" line in all the relevant files
    for current_file in files:
        with open(current_file,'r') as input_file:
            contents = input_file.readlines()
        found_subject_line = False
        for line in contents:
            if "subject:" in line.lower():
                found_subject_line = True
                break
        if not found_subject_line:
            print(f"{current_file} didn't have a subject line.")
            missing_subject_line_count += 1
        file_count += 1

print(f"examined a total of {file_count} files out of {total_files_count}, {missing_subject_line_count} of which didn't have a subject line.")
```

#### 1.2) populating twenty_newsgroup_dict_of_dicts_of_(topic_and_(topical_file_name)\_as_keys)\_and\_(file_valid_lines_as_values).pickle

```python
root_dir = '\\\\?\\' + os.path.abspath(os.path.join(os.getcwd(), '20_newsgroups'))

file_count = 0

subdirectory_list = ['alt.atheism', 'comp.graphics', 'comp.os.ms-windows.misc', 'comp.sys.ibm.pc.hardware', 'comp.sys.mac.hardware', 'comp.windows.x', 'misc.forsale', 'rec.autos', 'rec.motorcycles', 'rec.sport.baseball', 'rec.sport.hockey', 'sci.crypt', 'sci.electronics', 'sci.med', 'sci.space', 'soc.religion.christian', 'talk.politics.guns', 'talk.politics.mideast', 'talk.politics.misc', 'talk.religion.misc']

# iterating over every topic in the 20 newsgroups data set
for subdir in subdirectory_list:
    
    # finding the relevant files
    files = [ os.path.join(root_dir, subdir, f) for f in os.listdir(os.path.join(root_dir, subdir)) if '.onetoc2' not in f ] # '.onetoc2' is the file extension for Microsoft OneNote files, which tend to pollute subdirectories
    
    for current_file in files:
        
        lines_with_valid_content = []
        
        with open(current_file,'r') as input_file:
            
            contents = input_file.readlines()

        for i, line in enumerate(contents):
            
            # finding the "landmark" line in the header
            if "lines: " in line.lower():
                
                lines_with_valid_content.append(line.rstrip())
                
                break
        
        i += 1   
        
        # fetching all the lines after the "landmark" line
        while i < len(contents):
            
            lines_with_valid_content.append(contents[i].rstrip())
            
            i += 1
        
        subdirectory_dictionary = twenty_newsgroup_dict_of_dicts_of_topic_and_topical_file_name_as_keys_and_file_valid_lines_as_values[subdir]
        
        subdirectory_dictionary['{}+{}'.format(subdir, os.path.basename(current_file))] = lines_with_valid_content
        
        file_count += 1

print(f"examined a total of {file_count} files.")
```

#### 2.1) Using twenty_newsgroup_dict_of_dicts_of_topic_and_topical_file_name_as_keys_and_file_valid_lines_as_values to populate 20_newsgroups_dict_of_dicts_docID_token_freq_dicts

##### Helper Functions

```python
def has_some_alphanumeric_characters(line):
    '''
    Returns True/False depending on whether the input line (string) contains at least one alphanumeric character.
    
    Input:
        
        line: a string.
        
    Returns:
        boolean True/False
    '''
    if re.search('[a-zA-Z]', line):
        return True
    else:
        return False

def tokenize(list_of_lines, verbose=False):
    '''
    Flattens and tokenizes a list of lists (where each sublist is a single-element list representing a string that needs to be tokenized).
    
    Input:
        
        list_of_lines: a list of lists, where each sublist is a single-element list representing a string to tokenize.
        
        verbose: a boolean switch indicating the verboseness of this function.
    
    Returns: 
        
        tokenized_list: a list of all tokens with >= 1 alphanumeric character from the input list of lists
    '''
    tokenized_list = []
    for line in list_of_lines:
        tokens = [word for word in nltk.word_tokenize(line) if has_some_alphanumeric_characters(word)]
        if verbose:
            print('The line "{}" becomes:\n{}\n\n'.format(str(line), ', '.join(tokens)))
        for tok in tokens:
            tokenized_list.append(tok)
    return tokenized_list

# toy example
'''toy_example_list_of_valid_lines = ["NNTP-Posting-Host: punisher.caltech.edu",
                                   "sandvik@newton.apple.com (Kent Sandvik) writes:",
                                   ">>To borrow from philosophy, you don't truly understand the color red",
                                   ">>until you have seen it."] # taken from /alt.atheism/51128
#help(re.search)

tokenized_toy_example = tokenize(toy_example_list_of_valid_lines, verbose=True)

print(f"\nThe tokens in the toy example are: {', '.join(tokenized_toy_example)}")'''

# documentation on the Counter class
# help(Counter)

# example of using the Counter class
# tokens_and_frequencies = Counter(tokenized_toy_example)
# tokens_and_frequencies.most_common()
```

```python
twenty_newsgroups_dict_of_dicts_docID_token_freq_dicts = {}

for topic_name, topic_dictionary in twenty_newsgroup_dict_of_dicts_of_topic_and_topical_file_name_as_keys_and_file_valid_lines_as_values.items(): 
    # topic_name is one of 'alt.atheism', 'comp.graphics', ... 
    # and topic_dictionary is a filename::[list of valid lines in file] dictionary
    
    for filename, content_lines_in_filename in topic_dictionary.items():
        
        print(f"iterating over {topic_name} , {filename}")
        
        this_file_tokens_and_frequencies_dict = {} # this file's subdictionary
        
        twenty_newsgroups_dict_of_dicts_docID_token_freq_dicts[filename] = this_file_tokens_and_frequencies_dict
                            
        tokenized_list = tokenize(content_lines_in_filename)
        
        for token in tokenized_list:
            
            this_file_tokens_and_frequencies_dict[token] = tokenized_list.count(token)
        
```

#### 3) Making the twenty_newsgroups_corpus_wide_tokens_and_frequencies_(lexicon).pickle

```python
twenty_newsgroups_corpus_wide_tokens_and_frequencies_lexicon = None
twenty_newsgroups_corpus_wide_tokens = []

for topic_name, topic_dictionary in twenty_newsgroup_dict_of_dicts_of_topic_and_topical_file_name_as_keys_and_file_valid_lines_as_values.items(): 
    # topic_name is one of 'alt.atheism', 'comp.graphics', ... 
    # and topic_dictionary is a filename::[list of valid lines in file] dictionary
    
    for filename, content_lines_in_filename in topic_dictionary.items():
        
        print(f"iterating over {topic_name} , {filename}")
                                            
        tokenized_list = tokenize(content_lines_in_filename)
        
        twenty_newsgroups_corpus_wide_tokens.extend(tokenized_list)
        
twenty_newsgroups_corpus_wide_tokens_and_frequencies_lexicon = Counter(twenty_newsgroups_corpus_wide_tokens)
```
